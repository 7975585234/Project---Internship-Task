```python
# Import necessary libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import display, Image
import cv2
import os

# Construct the path to haarcascade_frontalface_default.xml
cascade_path = os.path.join(cv2.data.haarcascades, 'haarcascade_frontalface_default.xml')

# Create face cascade classifier
face_cascade = cv2.CascadeClassifier(cascade_path)
# Display the first image ('face4.jpg')
display(Image(filename=r'/a.jpg'))
# Load the image matrix from the first image
img = cv2.imread(r'/a.jpg')
# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Detect faces in the grayscale image
faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.05, minNeighbors=10)
# Iterate through the detected faces
for x, y, w, h in faces:
    # Extract the region of interest (ROI) - the detected face
    cropped_img = img[y:y+h, x:x+w]

    # Apply Canny edge detection to the cropped face
    edges = cv2.Canny(cropped_img, 130, 1000)

    # Count the number of edges in the cropped face
    number_of_edges = np.count_nonzero(edges)
# Check if the number of edges is greater than 1000
if number_of_edges > 1000:
    print(" no Wrinkle Found")
else:
    print(" Wrinkle Found")
```
